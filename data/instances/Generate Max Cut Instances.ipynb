```python
import networkx as nx
import pickle
import os
```

```python
# Generate 30 Erd√∂s Renyi graphs
for i in range(30):
    # Folder
    folder = "max_cut_" + str(i)
    
    # Check if folder exist
    if os.path.exists(folder) == False:
        # Create the directory
        os.mkdir(folder)
    
    # Generate Graph
    G = nx.generators.random_graphs.erdos_renyi_graph(n=8,p=0.5)
    
    path = folder + "/graph"
    # Save graph in Python pickle format.
    with open(path, "wb") as f:
        pickle.dump(G, f)
```

```python
# This is how you read files
G = pickle.load(open("max_cut_0/graph","rb"))
nx.draw_circular(G, node_size=1000, with_labels=True)
```
